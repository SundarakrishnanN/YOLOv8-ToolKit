```python
from ultralytics import YOLO
import os
import cv2

from PIL import Image
```

To train your YOLO model, the `dir` variable should point to the `data.yaml` file, which contains the dataset's path and configuration details. Ensure that this file is correctly set up to specify the locations of your training and validation data.

For more details on how to configure the `data.yaml` file, check the README of this repo.

The training hyperparameters need to be adjusted depending on what you are working on. Refer to the [Ultralytics YOLOv8 Documentation](https://docs.ultralytics.com/modes/train/).


```python

dir=''
model=YOLO("yolov8n.pt")
results = model.train(data=dir, epochs=300,dropout=0.3,label_smoothing=0.3,patience=50)
```

If you have an existing model and wish to log result metric,then execute the cell below. Specify the location of the model and the location of your `data.yaml` file. The validation dataset specified in your config file will be used to evaluate on.

Specify the location of your image and model and view the output of predictions on a single image.

```python
import matplotlib.pyplot as plt
model=YOLO("model path")

img_path ='Image path'
img = cv2.imread(img_path)
results=model(img)
r=results[0]
im_array = r.plot()  
im = Image.fromarray(im_array[..., ::-1])  
       
plt.imshow(im)
plt.show()
```
